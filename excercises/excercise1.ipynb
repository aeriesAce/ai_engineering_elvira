```python
import duckdb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/salaries.csv")

df.head(5)
```

```python
# b) plot the distribution of the salaries in USD

df_salaries_USD = duckdb.query(
    """
    SELECT 
        salary_in_usd
    FROM df
    
    """
).df()

sns.histplot(data=df_salaries_USD, x="salary_in_usd", bins=50)
plt.xlabel("Salary in USD")
```

```python
# c) How many job postings are there for each job title? 
# Take the tenth most common job title and plot a bar chart of the number of job postings for that title.
df_job_postings = duckdb.query(
    """
    SELECT 
        job_title,
        COUNT(*) AS postings
    FROM df
    GROUP BY job_title
    ORDER BY postings DESC
    LIMIT 1 OFFSET 9
    """
).df()

plt.barh(df_job_postings["job_title"], df_job_postings["postings"])
plt.xlabel("Tenth most common job title")
plt.ylabel("Postings")
```
