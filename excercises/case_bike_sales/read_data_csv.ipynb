```python
import pandas as pd

df = pd.read_csv("../../data/Sales.csv", index_col=0, parse_dates=True)
df.head()
```

```python
df.columns
```

```python
df.Sub_Category.sort_values(ascending=False)
```

```python
df.describe()
```

```python
print(df.value_counts())
```

```python
top_sale_per_year = (df.groupby(["Year", "Product"])["Profit"].sum().reset_index().sort_values(["Year", "Profit"], ascending=[True, False]).groupby("Year").head(5))
top_sale_per_year
```

```python
profit_by_age = df.groupby("Age_Group")["Profit"].sum().sort_values(ascending=False)
profit_by_age
```

```python
top_subcats = (
    df.groupby(["Age_Group", "Product"])
      .size()
      .reset_index(name="Antal_köp")
      .sort_values(["Age_Group", "Antal_köp"], ascending=[True, False])
)

top1_subcats = top_subcats.groupby("Age_Group").head(1)
top1_subcats
```

```python
top_profit_subcat = (
    df.groupby(["Age_Group", "Customer_Gender", "Sub_Category"])["Profit"].sum().reset_index().sort_values(["Age_Group", "Customer_Gender", "Profit"], ascending=[True, True, False]).head(1)
)

top1_profit_subcat = top_profit_subcat.groupby(["Age_Group", "Customer_Gender"]).head(1)
top1_profit_subcat
```

```python
top_cat_by_gender_age = (
    df.groupby(["Age_Group", "Customer_Gender", "Sub_Category"])["Order_Quantity"].sum().reset_index().sort_values(["Age_Group", "Customer_Gender", "Order_Quantity"], ascending=[True, True, False])
)

top1_cat_by_gender_age = top_cat_by_gender_age.groupby(["Age_Group", "Customer_Gender"]).head(1)
top1_cat_by_gender_age
```

```python
top_year_sales = (df.groupby(["Year", "Country", "Product_Category"])["Profit"].sum().reset_index().sort_values(["Year", "Profit"], ascending=[True, False]))
top5_year_sales = top_year_sales.groupby("Year").head(5)

top5_year_sales
```

```python
top_product_sale = (df.groupby(["Year", "Product_Category"])["Profit"].sum().reset_index())
top1_year_sales = top_product_sale.groupby("Year").head(1)
top1_year_sales
```

```python
total_sales = df.groupby(["Year", "Age_Group", "Country", "Product_Category", "Sub_Category", "Product"])[["Profit", "Order_Quantity"]].sum().sort_values(by="Profit", ascending=False).reset_index()
total_sales[total_sales["Product_Category"] == "Bikes"][total_sales["Country"] == "United States"]
```
