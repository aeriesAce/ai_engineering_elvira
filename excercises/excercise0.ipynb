```python
# b) Draw a line chart of quantity for each year. Is there a year that should be skipped?
df[["Year", "Quantity"]].groupby("Year").sum().reset_index()
```

```python
df[df["Year"] == 2017].sort_values(["Month"])

# if theres a year that should be skipped its 2017 since it's only one month worth of data in the file
```

```python
# first group the categories, then summarise the Quantity for each year
# drop the year 2017 from the df
df_line = df.groupby("Year")["Quantity"].sum().drop(2017)

# line chart with matplotlib
plt.plot(df_line)
plt.xlabel("Year")
plt.ylabel("Quantity")
```

```python
# c) Draw a line chart of average CO2 emissions for same years that as in b)
df_line_co2 = df.groupby("Year")["Avg_CO2"].sum().drop(2017)

plt.plot(df_line_co2)
plt.xlabel("Year")
plt.ylabel("Average CO2 emissions")
```

```python
# d) Draw a line chart of all years and months for import
# groupby year and month, summerise Import and reset the index
df_line_ym_import = df.groupby(["Year", "Month"])["Import"].sum().reset_index()

# Iplot with swaborn
plt.figure(figsize=(10,6))
sns.lineplot(data=df_line_ym_import, x="Month", y="Import", hue="Year", marker="o", legend="full")
plt.xticks(range(1, 12 + 1))
plt.title("Monthly imports by Year")

plt.show()
```
